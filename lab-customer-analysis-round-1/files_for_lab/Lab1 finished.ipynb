```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("file1.csv")
data2 = pd.read_csv("file2.csv")
data3 = pd.read_csv('file3.csv')
```

```python
data1.columns = data1.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
data2.columns = data2.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
data3.columns = data3.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
data2 = data2[data1.columns]
```
