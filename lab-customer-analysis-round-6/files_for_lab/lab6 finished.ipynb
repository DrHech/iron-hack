```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data = pd.read_csv("marketing_customer_analysis.csv")

data.describe()
```

```python
#Lab 5
y = data["Total Claim Amount"]
x = data.drop("Total Claim Amount", axis = 1).select_dtypes("number")
```

```python
from sklearn.preprocessing import Normalizer
transformer = Normalizer().fit(x)
x_normalized = transformer.transform(x)
```

```python
df_norm = pd.DataFrame(x_normalized)
```

```python
c = list(x.columns)
df_norm.columns = [c]
df_norm.head()
```

```python
#Lab 6 starts now:
str_data = data.select_dtypes(include = "object")
str_data = str_data.drop(columns = "Customer")
for col in str_data.columns:
    print("Dummification of the col: ",col)
    print(pd.get_dummies(str_data[col], drop_first = True))
```

```python
for col in str_data.columns:
    dummy = pd.get_dummies(str_data[col], drop_first = True)
    df_norm = pd.concat([df_norm,dummy], axis = 1)
df_norm = pd.concat([df_norm,data["Total Claim Amount"]], axis = 1)
df_norm.head()
```

```python
# Test Split
Y = df_norm["Total Claim Amount"]
X = df_norm.drop(columns = "Total Claim Amount")
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 100)

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from math import sqrt
lm = linear_model.LinearRegression()
model = lm.fit(X_train, Y_train)
lm.score(X,Y)
print ("lm intercept: ", lm.intercept_,"\n")
print ("lm coeff: ", lm.coef_,"\n")
predictions = lm.predict(X)


mae = mean_absolute_error(Y,predictions)
print ("mae is: ", mae, "\n")
mse = mean_squared_error(Y,predictions)
print ("mse is: ", mse, "\n")
rmse = sqrt(mse)
print ("rmse is: ", rmse, "\n")
r2 = r2_score(Y,predictions)
print ("r2 is: ", r2)
```
